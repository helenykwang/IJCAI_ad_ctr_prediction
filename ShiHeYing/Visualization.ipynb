```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
# %matplotlib notebook
%matplotlib inline
train=pd.read_csv(r'D:\Data\IJCAI\train_mispret.csv')#,na_values=-1 [160955, 205287, 214183, 314706]
test=pd.read_csv(r'D:\Data\IJCAI\test_mispret.csv')
col=['context_timestamp','num_click','num_conversion','context_id', 'user_id', 'item_id', 'shop_id', 'shop_score_delivery', 'shop_score_description', 'item_brand_id', 'item_sales_level', 'shop_score_service', 'shop_review_positive_rate', 'user_age_level', 'user_star_level', 'item_city_id', 'item_pv_level', 'shop_review_num_level', 'item_collected_level', 'item_category_list_0', 'context_page_id', 'item_price_level', 'user_occupation_id', 'user_gender_id', 'shop_star_level']
train=train[col]
test=test[col]
combine=[train,test]
for data in combine:
    data['timestamp']=pd.to_datetime(data['context_timestamp'],unit='s')
    data['day']=data['timestamp'].dt.day
train=pd.concat([train,test],axis=0)
```

```python
train=train.sort_values(by=['context_timestamp'])
train['context_timestamp']=train['context_timestamp']-1537200000
```

```python
plt.figure(figsize=(7,6))
sns.set(style="darkgrid")
ax = sns.countplot(x="day", hue='user_gender_id',data=train)
plt.figure(figsize=(16,6))
sns.set(style="darkgrid")
ax = sns.countplot(x="num_click", hue='user_gender_id',data=train, palette="Set3")
```
